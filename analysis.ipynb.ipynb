```python
import pandas as pd 
import numpy as np
import seaborn as sb
import matplotlib.pyplot as mp
```

```python
df=pd.read_excel("Churn_data.xlsx")
df
df.info()
```

```python
df["TotalCharges"]= df["TotalCharges"].replace (" ",0)
df.isnull()
```

```python
df.duplicated()
df.isnull()
df.isnull().sum()
df.describe()
```

```python
df.duplicated()
df['customerID'].duplicated().sum()
```

```python
def conv(value):
    if value==1:
        return "Yes"
    else:
        return"No"
    
df["SeniorCitizen"]= df["SeniorCitizen"].apply(conv)
df
```

```python
ax =sb.countplot(x="Churn", data=df, palette=["skyblue","orange"])
ax.bar_label(ax.containers[0],fontweight="bold")
ax.bar_label(ax.containers[1],fontweight="bold")
mp.title("COUNT OF CUSTOMERS BY CHURN", fontweight="bold",fontsize=16)
mp.figure (figsize=(4,4))
mp.show()
```

```python
gb=df.groupby("Churn").agg({"Churn":"count"})
gb
mp.pie(gb["Churn"],labels=gb.index, autopct="%1.2f%%",colors=["skyblue","orange"])
mp.title("PERCENTAGE CUSTOMERS BY CHURN", fontweight="bold",fontsize=14)
mp.show
```

```python
sb.countplot(x="gender",data=df, hue="Churn")
mp.title("ANALYZING GENDERS IMPACT ON CHURN", fontweight="bold",fontsize=14)
mp.show()
```

```python
sb.countplot(x="SeniorCitizen",data=df, hue="Churn")
mp.title("ANALYZING SENIOR CITIZEN IMPACT ON CHURN", fontweight="bold",fontsize=12)
mp.show()
```

```python
sb.histplot(x="tenure", data= df, bins=65)
mp.Figure(figsize=(6,4))
mp.title("CUSTOMER CHURN RATE BY TENURE", fontweight="bold",fontsize=14)
mp.show()
```

```python
ax =sb.countplot(x="Contract", data=df, hue="Churn")
ax.bar_label(ax.containers[0])
mp.title("COUNT OF CUSTOMERS BY CONTRACT")
mp.show()
```

```python

cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies']
num_cols = 3  
num_rows = (len(cols) + num_cols - 1) // num_cols  

fig, axes = mp.subplots(num_rows, num_cols, figsize=(15, 12))
axes = axes.flatten()  

for i, col in enumerate(cols):
    sb.countplot(data=df, x=col, ax=axes[i], palette='Set2', hue=df["Churn"])
    axes[i].set_title(f'Count Plot: {col}')
    axes[i].tick_params(axis='x')


for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

mp.tight_layout()
mp.show()
```

```python
ax =sb.countplot(x="PaymentMethod", data=df, hue="Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
mp.xticks( rotation =90)
mp.title("Churn customers by payment method")
mp.figure(figsize=(2,4))
mp.show()
```
